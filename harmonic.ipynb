```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
def symplectic_euler_step(q, p,h):
    p -= h*q
    q += h*p
    return (q,p)
```

```python
qs = []
ps = []
q = 1
p = 0
for i in range(1,1000):
    q,p = symplectic_euler_step(q,p,0.1)
    qs.append(q)
    ps.append(p)
```

```python
def energy(ps,qs,g=10):
    energies = []
    for p,q in zip(ps,qs):
        energies.append((p**2)/2)
    return energies
```

```python
matplotlib.rcParams['text.usetex'] = True
```

```python
plt.plot(qs,label = "$\\mathbf{q}$")
plt.plot(energy(qs,ps), label ="Energy")
plt.legend(fontsize = 15)
```
